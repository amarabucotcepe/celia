## 2. Bancos de dados

√â preciso configurar o driver do banco de dados pelo sqlalchemy para poder usar o m√©todo to_sql do pandas para salvar os dados no banco.

### 2.1. SQLite

Este exemplo usa um banco de dados sqlite local.

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///../data/bb.db', echo=True)
# sqlite:///relative/path/to/file.db
```

## 3. Salvar dados

### 3.1. Programa

Ap√≥s carregar os dados de programa de governo como um DataFrame √© poss√≠vel inserir a informa√ß√£o na tabela.

```python
programa = pd.DataFrame(r.values(), index=r.keys()).T
programa
```

Os dados de ag√™ncia e conta n√£o retornam pela api, ent√£o √© preciso registr√°-los no DataFrame para salvar no banco.

```python
programa['agency'] = agencia
programa['account'] = conta
```

```python
programa.to_sql('programa', con=engine, if_exists='replace')
```

### 3.2. Transa√ß√µes

Os dados de ag√™ncia e conta tamb√©m n√£o comp√µem o objeto transa√ß√£o, ent√£o √© necess√°rio registr√°-los.

```python
df['agency'] = agencia
df['account'] = conta
```

A chave prim√°ria da transa√ß√£o √© determinada pelo campo `referenceNumber`, e os campos `expensesCategory`, `expensesDocuments` s√£o m√∫ltiplos ent√£o s√£o exclu√≠dos do DataFrame antes da inser√ß√£o na tabela extrato.

```python
df.set_index('referenceNumber').drop(['expensesCategory','expensesDocuments'], axis=1).to_sql('extrato', con=engine, if_exists='replace')
```

### 3.3. Categoria

A informa√ß√£o da categoria √© extra√≠da do json por meio de uma desnormaliza√ß√£o com o m√©todo json_normalize que combinada ao campo `referenceNumber` permite criar uma tabela de categoria com chave estrangeira para a tabela de extrato.

```python
pd.json_normalize(transactions, 'expensesCategory', 'referenceNumber').to_sql('categoria', con=engine, if_exists='replace')
```

### 3.4. Documentos

A informa√ß√£o da documento √© extra√≠da do json por meio de uma desnormaliza√ß√£o com o m√©todo json_normalize que combinada ao campo `referenceNumber` permite criar uma tabela de documento com chave estrangeira para a tabela de extrato.

```python
pd.json_normalize(transactions, 'expensesDocuments', 'referenceNumber').to_sql('documentos', con=engine, if_exists='replace')
```
