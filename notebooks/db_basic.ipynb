## 2. Bancos de dados

Ã‰ preciso configurar o driver do banco de dados pelo sqlalchemy para poder usar o mÃ©todo to_sql do pandas para salvar os dados no banco.

### 2.1. SQLite

Este exemplo usa um banco de dados sqlite local.

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///../data/bb.db', echo=True)
# sqlite:///relative/path/to/file.db
```

## 3. Salvar dados

### 3.1. Programa

ApÃ³s carregar os dados de programa de governo como um DataFrame Ã© possÃ­vel inserir a informaÃ§Ã£o na tabela.

```python
programa = pd.DataFrame(r.values(), index=r.keys()).T
programa
```

Os dados de agÃªncia e conta nÃ£o retornam pela api, entÃ£o Ã© preciso registrÃ¡-los no DataFrame para salvar no banco.

```python
programa['agency'] = agencia
programa['account'] = conta
```

```python
programa.to_sql('programa', con=engine, if_exists='replace')
```

### 3.2. TransaÃ§Ãµes

Os dados de agÃªncia e conta tambÃ©m nÃ£o compÃµem o objeto transaÃ§Ã£o, entÃ£o Ã© necessÃ¡rio registrÃ¡-los.

```python
df['agency'] = agencia
df['account'] = conta
```

A chave primÃ¡ria da transaÃ§Ã£o Ã© determinada pelo campo `referenceNumber`, e os campos `expensesCategory`, `expensesDocuments` sÃ£o mÃºltiplos entÃ£o sÃ£o excluÃ­dos do DataFrame antes da inserÃ§Ã£o na tabela extrato.

```python
df.set_index('referenceNumber').drop(['expensesCategory','expensesDocuments'], axis=1).to_sql('extrato', con=engine, if_exists='replace')
```

### 3.3. Categoria

A informaÃ§Ã£o da categoria Ã© extraÃ­da do json por meio de uma desnormalizaÃ§Ã£o com o mÃ©todo json_normalize que combinada ao campo `referenceNumber` permite criar uma tabela de categoria com chave estrangeira para a tabela de extrato.

```python
pd.json_normalize(transactions, 'expensesCategory', 'referenceNumber').to_sql('categoria', con=engine, if_exists='replace')
```

### 3.4. Documentos

A informaÃ§Ã£o da documento Ã© extraÃ­da do json por meio de uma desnormalizaÃ§Ã£o com o mÃ©todo json_normalize que combinada ao campo `referenceNumber` permite criar uma tabela de documento com chave estrangeira para a tabela de extrato.

```python
pd.json_normalize(transactions, 'expensesDocuments', 'referenceNumber').to_sql('documentos', con=engine, if_exists='replace')
```
